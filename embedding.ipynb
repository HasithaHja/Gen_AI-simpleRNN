```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
# Sentences
sent = [
    "The cat sat on the mat.",
    "A dog chased the ball across the field.",
    "The boy reads a book every night before sleeping.",
    "She loves drinking coffee in the morning.",
    "The sun rises in the east and sets in the west.",
    "The scientist discovered a new planet in the galaxy.",
    "Music can change the way people feel."
]
```

```python
# Vocabulary size
voc_size = 10000
```

```python
# One hot encodeing
one_hot_repr = [one_hot(words, voc_size) for words in sent]
one_hot_repr
```

```python
# Embedding
import numpy as np
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
max_word_length = max([len(sentence.split()) for sentence in sent])
embedded_docs = pad_sequences(one_hot_repr, padding='pre', maxlen=max_word_length)
embedded_docs
```

```python
dim = 10
model = Sequential()
model.add(Embedding(voc_size, dim))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
